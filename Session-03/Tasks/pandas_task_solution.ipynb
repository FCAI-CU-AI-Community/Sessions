## Group by "City" and calculate the maximum "Total" for each "Product line"

```python
total_product_lines_each_city = data.groupby(["City", "Product line"])["Total"].sum()


cities = ["Mandalay", "Naypyitaw", "Yangon"]

for city in cities:
    highest_product_line = total_product_lines_each_city[city].idxmax()
    max_total = total_product_lines_each_city[city].max()
    print(f"In {city}, the product line with the highest total is '{highest_product_line}' with a total of {max_total}")

```
