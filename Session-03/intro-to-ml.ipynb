## Exploring and Visualizing Data: Pandas and Matplotlib

### Pandas
- resources:
    [Hands on ML - Pandas Tutorial](https://colab.research.google.com/github/ageron/handson-ml3/blob/main/tools_pandas.ipynb)

#### Why?
**Pandas** is used because it offers a powerful, flexible, and efficient way to manipulate, analyze, clean, and visualize data, which is essential for any data science or machine learning task.

#### 1. Data Structures:
- **Series**: A one-dimensional labeled array, similar to a list or column in a spreadsheet. Each element in a series has an associated label or index.
- **DataFrame**: A two-dimensional, tabular data structure with labeled axes (rows and columns). It’s comparable to a spreadsheet or SQL table, but more powerful in terms of operations.

![ML Development Cycle](blobs/pandas1.png)


#### 2. Important Features
- Data Importing and Exporting (csv, json, sql, excel)
- Data Cleaning
- Data manipulation (indexing, selection, filtering, merge/join)
- Calculate Statistics (mean, median, sum, min, max)
- Handling Dates and Times
- Group by and Aggregation
- Pivoting and Reshaping Data

Matplotlib’s primary purpose is to provide users with the tools and functionality to represent data graphically, making it easier to analyze and understand.
