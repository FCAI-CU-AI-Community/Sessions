```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

# Parameters
n_points = 25  # number of data points
noise_level = 5  # noise level

# Generate random x values
x = np.linspace(0, 10, n_points)

# Generate random y values
def generate_data(m, c):
    y = m * x + c + np.random.normal(0, noise_level, n_points)
    return y

# Define a plotting function that takes m, c, marker shape, and line thickness as parameters
def plot_line(w=1.0, b=0.0):
    y = generate_data(2.5, 5)
    plt.figure(figsize=(5, 5))
    plt.scatter(x, y, color='r', label="Our data", marker='x')
    plt.plot(x, w * x + b, color='#8ab7e2', label="Linear model", linewidth=3)
    
    # Add dashed lines from each data point to the line
    for xi, yi in zip(x, y):
        plt.plot([xi, xi], [yi, w * xi + b], 'b--', linewidth=0.8)  # Dashed lines
    
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.title("Linear Regression Model")
    plt.grid(True)
    plt.show()

# Create interactive sliders for m, c, a dropdown for marker shape, and line thickness
interact(
    plot_line,
    w=widgets.FloatSlider(min=-10, max=10, step=0.1, value=2.5),
    b=widgets.FloatSlider(min=-20, max=20, step=0.5, value=5),
)
```

```python
# Parameters
n_points = 25  # number of data points
noise_level = 5  # noise level
slope_at_point = 2 * -6
# Generate random x values
x = np.linspace(-10, 10, n_points)
y = x ** 2

plt.figure(figsize=(5, 5))
plt.scatter(-6, 6**2, color='r', marker='o', s=100)
plt.scatter(6, 6**2, color='r', marker='o', s=100)
plt.scatter(5, 5**2, color='g', marker='o', s=100)
plt.scatter(-5, 5**2, color='g', marker='o', s=100)

plt.plot(x, y, color='#8ab7e2', linewidth=3)
# Plot a tangent line at the specific point
tangent_x_neg = np.array([-6 - 6, -6 + 6])  # A small range around the specific point
tangent_y_neg = slope_at_point * (tangent_x_neg + 6) + (6 ** 2)  # y = mx + b
# Plot a tangent line at the specific point
tangent_x_pos = np.array([6 - 6, 6 + 6])  # A small range around the specific point
tangent_y_pos = -slope_at_point * (tangent_x_pos - 6) + (6 ** 2)  # y = mx + b
plt.plot(tangent_x_neg, tangent_y_neg, color='orange', label=f'Slope of Curve at point 6, equal {slope_at_point}', linewidth=2, linestyle='--')
plt.plot(tangent_x_pos, tangent_y_pos, color='green', label=f'Slope of Curve at point 6, equal {-slope_at_point}', linewidth=2, linestyle='--')
plt.xlabel("w")
plt.ylabel("J(w)")
plt.title("Loss Function")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data with extended range
np.random.seed(42)
x = np.linspace(-10, 10, 20)
y = 0.4 * x + np.random.normal(0, 0.8, size=len(x))  # Increased noise to match scale

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# First subplot: Data and fit
thetas = [0.000, 0.184, 0.301, 0.374, 0.420]
colors = ['blue', 'red', 'green', 'cyan', 'orange']

# Plot the scattered points
ax1.scatter(x, y, color='black', marker='x')

# Plot different theta lines
x_line = np.linspace(-10, 10, 100)
for theta, color in zip(thetas, colors):
    y_line = theta * x_line
    ax1.plot(x_line, y_line, color=color, label=f'w={theta:.3f}')

ax1.set_xlim(-10, 10)
ax1.set_ylim(-5, 5)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title('Data and fit')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Second subplot: Cost function
theta_range = np.linspace(-0.2, 1.0, 100)
cost = np.zeros_like(theta_range)

# Calculate cost function (MSE)
for i, theta in enumerate(theta_range):
    predictions = theta * x
    cost[i] = np.mean((predictions - y) ** 2) / 2

ax2.plot(theta_range, cost, 'k-')

# Plot points corresponding to the thetas from first plot
for theta, color in zip(thetas, colors):
    pred = theta * x
    current_cost = np.mean((pred - y) ** 2) / 2
    ax2.plot(theta, current_cost, 'o', color=color)

ax2.set_xlim(-0.2, 1.0)
ax2.set_ylim(-0.5, 15)  # Adjusted y-range for the new scale
ax2.set_xlabel('w')
ax2.set_ylabel('J(w)')
ax2.set_title('Cost function')
ax2.grid(True, alpha=0.3)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
# plt.style.use('seaborn')

# Generate sample data
np.random.seed(42)
x = np.linspace(-2, 2, 100)
true_slope = 198.5
true_intercept = -197.8
y = true_slope * x + true_intercept + np.random.normal(0, 50, size=len(x))

# Create figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Setup the first subplot
scatter = ax1.scatter(x, y, color='blue', alpha=0.5, label='Data points')
line, = ax1.plot([], [], 'r-', label='Regression line')
residuals = ax1.vlines(x, y, y, colors='red', alpha=0.3)
mse_text = ax1.text(0.02, 0.85, '', transform=ax1.transAxes)

ax1.set_xlim(-2.5, 2.5)
ax1.set_ylim(-700, 200)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Setup the second subplot - now with quadratic MSE
param_range = np.linspace(-200, 600, 200)  # wider range around true slope
mse_values = np.zeros_like(param_range)
for i, a in enumerate(param_range):
    # Calculate MSE for each parameter value
    y_pred = a * x + true_intercept
    mse_values[i] = np.mean((y - y_pred) ** 2)

mse_line, = ax2.plot(param_range, mse_values, 'b-')
current_mse, = ax2.plot([], [], 'ro')

ax2.set_xlim(-200, 600)
ax2.set_ylim(0, np.max(mse_values) * 1.1)
ax2.set_xlabel('w')
ax2.set_ylabel('J(w)')
ax2.set_title('J(w) vs parameter w')

# Animation function
def animate(frame):
    # Current parameter value - sweep through parameter range
    frame_progress = frame / 200  # normalize to 0-1
    # Use sinusoidal motion for smoother animation
    current_slope = param_range[0] + (param_range[-1] - param_range[0]) * (1 + np.cos(2 * np.pi * frame_progress)) / 2
    
    # Update regression line
    y_pred = current_slope * x + true_intercept
    line.set_data(x, y_pred)
    
    # Update residuals
    segments = np.c_[x, y, x, y_pred].reshape(-1, 2, 2)
    residuals.set_segments(segments)
    
    # Calculate and update MSE
    current_mse_value = np.mean((y - y_pred) ** 2)
    mse_text.set_text(f'MSE = {current_mse_value:.1f}\ny = {current_slope:.1f}x + {true_intercept:.1f}')
    
    # Update MSE point
    current_mse.set_data([current_slope], [current_mse_value])
    
    return line, residuals, mse_text, current_mse

# Create animation
anim = FuncAnimation(fig, animate, frames=200, interval=50, blit=True)

# Save as GIF
writer = PillowWriter(fps=20)
plt.tight_layout()
anim.save('regression_animation.gif', writer=writer)
plt.close()

print("Animation has been saved as 'regression_animation.gif'")
```
